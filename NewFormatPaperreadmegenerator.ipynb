```python
def removeSpaces(string):
 
    
    count = 0
 
    list = []
 
  
    for i in range(len(string)):
        if string[i] != ' ':
            list.append(string[i])
 
    return toString(list)
 
# Utility Function
def toString(List):
    return ''.join(List)
```

```python
def stringClean(String1):
    
    val1 =re.sub(",","",String1)
    val2 =re.sub("-","",val1)
    val3 =re.sub(":","",val2)
    
    return val3
```
